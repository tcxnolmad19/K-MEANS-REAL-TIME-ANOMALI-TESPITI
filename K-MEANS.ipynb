# model parametre bulma

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd
import gc  # Garbage Collector


# 1. Veriyi Hazırlama ve Kategorik Sütunları Sayısal Hale Dönüştürme
df_encoded = df_set5.copy()

# Kategorik sütunları Label Encoding ile sayısal hale dönüştürme
label_encoders = {}
for column in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[column] = le.fit_transform(df_encoded[column])
    label_encoders[column] = le  # Encoder'ları saklamak için

# Veriyi eğitim ve test setlerine bölme
df_train, df_test = train_test_split(
    df_encoded,
    test_size=0.2,
    random_state=42,
    stratify=df_encoded['class']
)

# 2. Veriyi Ölçeklendirme
scaler = StandardScaler()
df_train_scaled = scaler.fit_transform(df_train.drop('class', axis=1))
df_test_scaled = scaler.transform(df_test.drop('class', axis=1))

# 3. Random Forest ile Özellik Önemini Belirleme
rf = RandomForestClassifier(n_estimators=17, random_state=42)
rf.fit(df_train_scaled, df_train['class'].replace({'normal': 0, 'anomaly': 1}))

# Özellik önemlerini alma
feature_importances = pd.Series(rf.feature_importances_, index=df_train.drop('class', axis=1).columns)
sorted_features = feature_importances.sort_values(ascending=False).index

print("Özelliklerin önem sırasına göre listesi:")
print(feature_importances.sort_values(ascending=False))

# 4. En İyi Sonucu Elde Etmek için Tüm Kombinasyonları Deneme
best_accuracy = 0
best_params = {}
best_selected_features = []

# Farklı `k` değerleri, eşik değerleri ve özellik sayısı ile optimizasyon döngüsü
for n_features in range(1, len(sorted_features) + 1):
    selected_features = sorted_features[:n_features]
    df_train_selected = df_train[selected_features].copy()
    df_test_selected = df_test[selected_features].copy()

    # Veriyi ölçeklendirme 
    scaler_selected = StandardScaler()
    df_train_scaled_selected = scaler_selected.fit_transform(df_train_selected)
    df_test_scaled_selected = scaler_selected.transform(df_test_selected)

    for k in range(2, 25):  # Farklı `k` değerleri
        kmeans = KMeans(n_clusters=k, n_init=20, max_iter=50, random_state=42)
        kmeans.fit(df_train_scaled_selected)

        # Eğitim ve test verilerinin en yakın küme merkezine olan mesafelerini hesaplama
        train_distances = kmeans.transform(df_train_scaled_selected).min(axis=1)
        test_distances = kmeans.transform(df_test_scaled_selected).min(axis=1)

        for threshold_percentile in range(20, 70, 1):  # Farklı eşik dilimleri
            threshold = np.percentile(train_distances, threshold_percentile)
            y_pred = (test_distances > threshold).astype(int)
            y_true = df_test['class'].replace({'normal': 0, 'anomaly': 1}).values

            accuracy = accuracy_score(y_true, y_pred)

            # En yüksek doğruluğu güncelleme
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_params = {
                    'n_features': n_features,
                    'k': k,
                    'threshold_percentile': threshold_percentile,
                    'accuracy': accuracy
                }
                best_selected_features = selected_features.tolist()


print("\nEn iyi parametreler:")
print(f"Özellik Sayısı: {best_params['n_features']}")
print(f"K Değeri: {best_params['k']}")
print(f"Eşik Değeri Yüzdelik Dilimi: {best_params['threshold_percentile']}")
print(f"En İyi Accuracy: {best_params['accuracy']}\n")

print("En iyi parametrelerle seçilen özellikler:")
for feature in best_selected_features:
    print(f"- {feature}")

del df_encoded
del df_train
del df_test
del df_train_selected
del df_test_selected

gc.collect()

print("\nGereksiz DataFrame'ler silindi ve bellek temizlendi.")
```

```python

import pandas as pd
import numpy as np
import os
import time
from io import StringIO

# -------------------------
# 1. Anomali Tespiti ve İzleme Fonksiyonu
# -------------------------
def monitor_file(file_path, chunk_size=500, pause_time=10, anomaly_ratio_threshold=0.8):
  
    # Başlangıçta işlenen satır sayısını 0 olarak ayarla
    processed_rows = 0
    buffer = []

    while True:
        if not os.path.exists(file_path):
            print(f"Dosya bulunamadı: {file_path}")
            time.sleep(pause_time)
            continue

        try:
            with open(file_path, 'r') as f:
                # İlk satırda CSV başlıklarını (header) okuyoruz
                header = f.readline().strip().split(',')
                
                src_ip_idx = header.index('src_ip')
                dst_ip_idx = header.index('dst_ip')
                
                # Daha önce işlenmiş satırları atla
                for _ in range(processed_rows):
                    f.readline()

                # Yeni gelen satırları oku
                new_lines = f.readlines()

                if not new_lines:
                    # Yeni satır yoksa bekleyerek döngünün devam etmesini sağla
                    time.sleep(pause_time)
                    continue

                # Yeni satırları buffer'a aktar
                buffer.extend(new_lines)
                processed_rows += len(new_lines)

                # Buffer’da yeterli satır varsa (örneğin 500 satır) işlemeye başla
                while len(buffer) >= chunk_size:
                    # Alınacak 500 (chunk_size) satır
                    chunk_lines = buffer[:chunk_size]
                    # Buffer'dan bu satırları çıkart
                    buffer = buffer[chunk_size:]

                    # Okunan satırları DataFrame'e dönüştür
                    df_chunk = pd.read_csv(
                        StringIO(''.join(chunk_lines)),
                        header=None,      # Çünkü başlığı ayrı okuduk
                        names=header      # Daha önce okuduğumuz başlık isimlerini kullan
                    )

                    # Model için gerekli dönüşümleri yapacak DataFrame
                    df_new = df_chunk.copy()

                   
                    # Örnek: LabelEncoder ve benzeri dönüşümler
                    for column in selected_features:
                        if column in label_encoders:
                            df_new[column] = df_new[column].astype(str)
                            try:
                                df_new[column] = label_encoders[column].transform(df_new[column])
                            except ValueError as e:
                                print(f"Hata: {column} sütununda bilinmeyen kategori tespit edildi. Satır(lar) atlanıyor.")
                                df_new = df_new.dropna(subset=[column])

                    # Modelde kullanılacak sütunları seç
                    df_new_selected = df_new[selected_features].copy()

                    # Ölçeklendirme
                    df_new_scaled = scaler_sel.transform(df_new_selected)

                    # K-Means üzerinden anomali tespiti
                    new_distances = kmeans.transform(df_new_scaled).min(axis=1)
                    new_predictions = (new_distances > threshold).astype(int)
                    num_anomaly_new = np.sum(new_predictions)

                    # Orijinal DataFrame'e (df_chunk) 'prediction' sütunu ekle
                    df_chunk['prediction'] = new_predictions
                    df_chunk['prediction'] = df_chunk['prediction'].replace({0: 'normal', 1: 'anomaly'})

                    # Kaç normal / anomali var
                    print(f"\nBu 500'lü bloktaki (Toplam Satır: {len(df_chunk)}) tahminler:")
                    print(df_chunk[['prediction']])
                    print(f"\nNormal sayısı: {len(df_chunk) - num_anomaly_new}")
                    print(f"Anomaly sayısı: {num_anomaly_new}")

                    # Anomali oranı
                    anomaly_ratio = num_anomaly_new / len(df_chunk)

                    # Blokta anomali oranı threshold'u geçerse
                    if anomaly_ratio > anomaly_ratio_threshold:
                        # Anormal satırları (anomali) çek
                        anomalies = df_chunk[df_chunk['prediction'] == 'anomaly']

                        print(f"\nUYARI: Bu 500'lü blokta anomali oranı = {anomaly_ratio:.2f} > {anomaly_ratio_threshold}")
                        print("Gerçek zamanlı bildirim gönderiliyor...\n")

                        # --------------------------------------------------------
                        # Gerçek satır indeksini hesaplamak
                        # --------------------------------------------------------
                        
                        chunk_base_index = (processed_rows - len(buffer) - chunk_size)

                        # Anomalik satırlar için src_ip, dst_ip ve gerçek satır indeksini yaz
                        for i, row in anomalies.iterrows():
                            real_line = chunk_base_index + i + 1
                            print(
                                f"GERÇEK ZAMANLI BİLDİRİM => "
                                f"Satır Index (dosyadaki): {real_line}, "
                                f"src_ip: {row['src_ip']}, "
                                f"dest_ip: {row['dst_ip']}, "
                                f"Durum: ANOMALY"
                            )

                    # Anomali sayısı > 0 ise genel bir uyarı
                    if num_anomaly_new > 0:
                        print(f"UYARI: Bu blokta {num_anomaly_new} anomali (saldırı) tespit edildi!")

        except Exception as e:
            print(f"Dosya okunurken hata oluştu: {e}")
            time.sleep(pause_time)
            continue

# -------------------------
# 2. İzleme İşlemini Başlatma
# -------------------------
file_path = '/home/selcuk1453/attack/network_features.csv'  # İzlenecek dosyanın yolu



monitor_file(
    file_path=file_path,
    chunk_size=500,
    pause_time=10,
    anomaly_ratio_threshold=0.8  # 80% üzeri anomali çıkarsa bildirim
)
```
